## Unit of Analysis

One row represents the daily performance of one content item for one client on one report date.

## Time Window

The warehouse dataset covers data from 2025-01-27 to 2026-06-30 (approximately 17 months).
The analysis should use historical data available before the prediction point.

```python
SELECT
    COUNT(*) AS total_rows,
    MIN(report_date) AS start_date,
    MAX(report_date) AS end_date
FROM fact_content_daily_performance;
```

### Features

- gsc_impressions
- gsc_clicks
- gsc_avg_position
- client_has_gsc
- client_has_ga4

These are known before prediction.

### Label

Future performance metric (target variable).

### Context

- client_hash_id
- content_hash_id
- report_date

These identify the observation and should not be used as model features.

### Excluded

Future metrics, target-derived variables and any private identifiers.

```python
SELECT
    COUNT(gsc_impressions) AS impressions,
    COUNT(gsc_clicks) AS clicks,
    COUNT(gsc_avg_position) AS positions
FROM fact_content_daily_performance;
```

The grain is one row per client, content item and report date.

The dataset should not contain duplicate rows for the same grain.

```python
SELECT
    report_date,
    client_hash_id,
    content_hash_id,
    COUNT(*) AS duplicates
FROM fact_content_daily_performance
GROUP BY
    report_date,
    client_hash_id,
    content_hash_id
HAVING COUNT(*) > 1
LIMIT 5;
```

Missing values are expected because not every client has GA4 or GSC connected.

Missingness is likely related to client configuration rather than random data loss.

```python
SELECT
AVG(CASE WHEN gsc_impressions IS NULL THEN 1 ELSE 0 END) AS missing_impressions,
AVG(CASE WHEN gsc_clicks IS NULL THEN 1 ELSE 0 END) AS missing_clicks,
AVG(CASE WHEN gsc_avg_position IS NULL THEN 1 ELSE 0 END) AS missing_position
FROM fact_content_daily_performance;
```
